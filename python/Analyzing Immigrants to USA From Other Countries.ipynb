# About Dataset

Dataset: This data set contains immigration data to United States of America from 1980 to 2013 - International migration flows to and from selected countries - The 2015 revision from United Nation's website.

It contains yearly data on the flows of international migrants as recorded by the countries of destination. The data shows inflows according to the place of birth, citizenship or place of previous / next residence both for foreigners and nationals. 

You can find the dataset in repository.


```python
import numpy as np  # useful for many scientific computing in Python
import pandas as pd # primary data structure library
```

```python
df_us = pd.read_excel('USA.xlsx',
                       sheet_name='USA by Place of birth')

print ('Data read into a pandas dataframe!')
```

```python
df_us.head()
```

```python
# Add Total Column
df_us['Total'] = df_us.sum(axis=1)
```

```python
# Rename Columns
df_us.rename(columns={'OdName':'Country', 'AreaName':'Continent', 'RegName':'Region'}, inplace=True)
df_us.head(2)
```

```python
# Check whether type of all columns are correct
df_us.info()
```

```python
df_us.describe()
```

```python
#Set Country as index for our DS
df_us.set_index('Country', inplace=True)
```

```python
df_us.head()
```

```python
years = list(map(str, range(1980, 2014)))
```

```python
# use the inline backend to generate the plots within the browser
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # optional: for ggplot-like style

# check for latest version of Matplotlib
print ('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

# Trend for Top 5 Countries

In this section I find the top 5 Countries that contributed most to Immigration to USA.

To find the top 5 countries, I use **Total** column which I added in above steps. 
I sort data by using **Total** column in descending order, then show the trend using line graph.

```python
df_us.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_us.head()
```

```python
df_top5 = df_us.head(5)
df_top5 = df_top5[years].transpose() 
df_top5.index = df_top5.index.map(int)
df_top5.plot(kind='line', figsize=(13, 8))

plt.title('Immigration Trend of Top 5 Countries')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

# Analyze Mexico Data Deeply

In above graph you can see there is a sudden peek around 1989 to 1991. There is a need to analyze data in detail. Therefore I plot bar graph.


```python
df_mexico = df_us.loc['Mexico', years]
df_mexico.head()
```

```python
df_mexico.plot(kind='bar', figsize=(10, 6), rot=90) 

plt.xlabel('Year') # add to x-label to the plot
plt.ylabel('Number of immigrants') # add y-label to the plot
plt.title('Mexican immigrants to USA from 1980 to 2013') # add title to the plot

# Annotate arrow
plt.annotate('',                             # s: str. will leave it blank for no text
             xy=(10.5, 960000),             # place head of the arrow at point (year 1991 )
             xytext=(8.5, 420000),         # place base of the arrow at point (year 1989 )
             xycoords='data',         # will use the coordinate system of the object being annotated 
             arrowprops=dict(arrowstyle='->', connectionstyle='arc3', color='blue', lw=2)
            )

# Annotate Text
plt.annotate('IRCA 1986', # text to display
             xy=(8.5, 600000),                    # start the text at at point (year 1989 )
             rotation=80,                  # based on trial and error to match the arrow
             va='bottom',                    # want the text to be vertically 'bottom' aligned
             ha='left',                      # want the text to be horizontally 'left' algned.
            )

plt.show()
```

I find some interesing insights. During 1980s Mexican migration had increased dramatically. The Immigration Reform and Control Act (IRCA) of 1986 granted amnesty to illegal immigrants who had resided in the U.S. before 1982.<sup><a href="https://en.wikipedia.org/wiki/History_of_Mexican_Americans#Since_1965">[1]</a></sup>

The spike in the numbers of new immigrants from 1989 to 1991 does not represent a surge of new arrivals but rather the change in legal status. May be due to IRCA 1986.<sup><a href="https://www.nap.edu/read/23550/chapter/5">[2]</a></sup>

# Finding Trend between top 3 Continents

In this segment I find trend between top 3 continents that contributed the most to immigration to USA.

I reset previous index of the data set and make index on **Continent** column, then I make groups by using continents. After that I show the trend using line graph.

```python
df_us_without_index = df_us[df_us.notnull()].copy()
df_us_without_index.reset_index(inplace=True)
df_us_without_index.head()
```

```python
# Create new list to get Continent's, Total and years data
years_with_more_cols = years + ['Continent', 'Total']
df_us_continent_group = df_us_without_index[years_with_more_cols]
df_us_continent_group = df_us_continent_group.groupby(['Continent'],as_index=False).sum()
df_us_continent_group.sort_values(by='Total', ascending=False, axis=0, inplace=True)
df_us_continent_group.set_index('Continent', inplace=True)
```

```python
df_top_cont3 = df_us_continent_group.head(3)
df_top_cont3 = df_top_cont3[years].transpose() 
df_top_cont3.index = df_top_cont3.index.map(int)
df_top_cont3.plot(kind='line', figsize=(13, 8))

plt.title('Immigration Trend of Top 3 Continents')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```

# Show Top country among Top 3 Continents

In the previous section you saw I find top 3 continents. In the following portion I find the top country among the top 3 Continents.

First of all I extract list of values of top 3 continent, then using that list I make the list of names of the country in each continent. After that get those countries and finally show the trend between those countries.

```python
# Get top 3 continents and convert to python list
top3_continents_lst = df_us_continent_group.head(3).index.values.tolist()
df_us_top_countries = df_us_without_index[df_us_without_index.notnull()].copy()
df_us_top_countries.set_index('Continent', inplace=True)
df_us_top_countries.sort_values(by='Total', ascending=False, axis=0, inplace=True)
```

```python
#Get Top country which contributed most with Top 3 Continent
top_us_most_countries =  [df_us_top_countries.loc[r].iloc[0].Country for r in top3_continents_lst] 

#Get Data set of top 3 countries in top 3 Continents
df_us_top3_cont_countries = df_us.loc[top_us_most_countries]
df_us_top3_cont_countries
```

```python
df_us_top3_cont_countries = df_us_top3_cont_countries[years].transpose() 
df_us_top3_cont_countries.index = df_us_top3_cont_countries.index.map(int)
df_us_top3_cont_countries.plot(kind='line', figsize=(13, 8))

plt.title('Immigration Trend of Top Country among Top 3 Continents')
plt.ylabel('Number of Immigrants')
plt.xlabel('Years')
plt.show()
```
